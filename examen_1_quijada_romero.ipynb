1. Escriba un código que sume los números múltiplos de 13 que hay entre 0 y 10,000, es decir, que sume 0 + 13 + 26 + 39 + ... + 9,997.

````python
suma = 0 #Valor inicial de la suma
for i in range(0,10001): #Toma los números entre [0,10001) = [0,10000] (Tomamos uno más para considerar al 10'000)
    if i%13 == 0: #Si es múltiplo de 13
        suma += i #Sumamos el valor
print(suma) #Imprimimos el resultado
````

2. Haga la gráfica de la función $f(x)=x + \frac{x^3}{3!} + \frac{x^5}{5!}$ y de la función $\sinh (x)$ en la misma gráfica. Como se comparan las dos funciones anteriores en el intervalo de $-1$ a $1$. Explique las similitudes y diferencias.

````python
dom = linspace(-1,1,1000) #Damos el dominio con 1000 puntos
f = dom + (dom**3)/(math.factorial(3)) + (dom**5)/(math.factorial(5)) #Definimos la función f (que es la expansión en serie de Taylor)
senh = sinh(dom)#Damos los puntos para el seno hiperbólico

fig, ax = plt.subplots() #Se definen la figura y los ejes
plot(dom, f, '-', linewidth = 2, label = "Taylor") #Se grafica la función f
plot(dom, senh, '--', label = "Biblioteca") #Se grafica la función sinh
ax.set(title = 'Gráficas de la función f(x) y senh(x)') #Definición del título
ax.set(xlabel = 'x', ylabel = 'f(x)') #Definición de los ejes
grid() #Se muestra la lattice 
legend() #Se muestra cuál es cuál
show() #Muestra de la gráfica
````

````python
dif = abs(senh-f) #Nueva lista con la diferencia entre los valores de ambas funciones
print(max(dif)) #Se obtiene el máximo de las diferencias
````

A primera vista se observa que las gráficas coinciden en todo el dominio considerado; la función $f$ se sobrepone a la función dada por la biblioteca importada. Una observación inmediata es que ambas son estrictamente crecientes en todo el intervalo. Además, ambas pasan por el 0

Por otro lado, de acuerdo con el bloque anterior de código, la diferencia máxima entre ambas funciones es del orden de $10^{-3}$.

4. Realice una función que reciba tu número de cuenta como entero y que a la salida proporcione la suma de todos sus dígitos.

````python
def numtosum(cuenta = 317163670): #Definimos la función con mi número de cuenta como valor de la entrada por default
    suma = 0 #Suma
    for i in str(cuenta): #Convierte el número a string e itera por sus caracteres
        suma += int(i) #Convierte a entero y suma el valor a la suma
    return suma

print(numtosum()) #Prueba: Da el valor correcto
````
